# Part 1: Underwater Acoustics (Lecture Questions)

```python
speed_of_sound_baltic = 1495  # Speed of sound in Baltic Sea water in m/s

def calculate_wavelength(frequency_khz, speed_of_sound_ms):
    """Calculate the wavelength of a sound wave in water."""
    frequency_hz = frequency_khz * 1000  # Convert kHz to Hz
    wavelength_m = speed_of_sound_ms / frequency_hz
    return wavelength_m

frequencies_khz = [120, 38]

# Print the results
for d, s in zip(frequencies_khz, [calculate_wavelength(f, speed_of_sound_baltic) for f in frequencies_khz]):
    print(f"Frequency: {d} kHz, Wavelength: {s * 100:.2f} centimeters")
```

```python
rho_baltic = 1025        # Density of Baltic Sea water in kg/m^3
p_source = 200*10**3     # Sound pressure in Pa
p_ref = 1e-6             # Reference sound pressure in µPa

# dB = 10 * log10(I / I_ref) = 20 * log10(P / P_ref)
dB = 20 * math.log10(p_source / p_ref)  

print(f"Decibel Level: {dB:.2f} dB")
```

```python
# Decibels reduced by 6 dB 
new_dB = dB - 6

# solve for new pressure level p_source_new = p_ref * 10**(new_dB / 20)
p_source_new = p_ref * 10**(new_dB / 20)

print(f"New Pressure Level: {p_source_new / 1000:.2f} kPa")
print(f"Ratio of new pressure to original pressure: {p_source_new / p_source:.2f}")
```

```python
I_0, I_1 = sp.symbols('I_0 I_1')

sound_intensity = 0.001 / 100       # Sound intensity in procent
I_1 = I_0 * sound_intensity         # I_1 is the reflected intensity, I_0 is the emitted intensity


# Calculate the energy flux in W/m^2
I = 10 * math.log10(I_0/I_1)  # Convert promille to W/m^2 and calculate dB

print("\nTASK 4:")
print(f"Change in dB: {I:.2f} dB")
```

# Part 2: Propagation

1. Source depth is 200 m. Horizontal source location is the origin of the range axis (r = 0 km) in this 2D run.

3. Water depth is 5000 m. The SSP is a deep sound channel (Munk-like): sound speed decreases from about 1548.5 m/s at surface to a minimum about 1500.1 m/s near 1400 m, then increases again to about 1551.9 m/s at 5000 m. So it is a U-shaped profile with a minimum-speed axis.

4. Surface boundary: VACUUM (pressure-release surface, strong phase-reversing reflection idealization). Bottom boundary: acousto-elastic half-space at 5000 m with: compressional speed about 1600 m/s, density about 1.80 g/cm^3, attenuation parameter alphaI = 0.8000 (with attenuation units reported as dB/m). Water-column attenuation includes THORP volume attenuation.

5. At the minimum of the SSP, approximately 1400 m (around 1200 to 1600 m, minimum value at 1400 m in your table).

6. Rays launched at relatively small grazing angles around horizontal are trapped in the channel (they refract toward and oscillate around the axis instead of immediately hitting boundaries). In your run (−12° to +12°), this is typically the central part of the fan (angles near 0°, moderate magnitudes), while steeper angles are more likely to interact with surface or bottom.

7. Because of refraction in the U-shaped SSP. Above the axis, negative gradient bends rays downward; below the axis, positive gradient bends rays upward. This creates turning points, so many rays cycle around the channel axis and avoid both boundaries.

8. Convergence zones: at ranges where refracted ray families reconverge after one or more channel cycles (in deep-channel environments, these are commonly periodic at long range). Shadow zones: regions between ray bundles where no eigenrays pass (low insonification), often between convergence regions and at depths/ranges not intersected by the traced fan. Exact CZ/shadow locations require the ray plot or transmission-loss plot from this run, but physically they are organized around the channeling structure centered near the ~1400 m axis.
