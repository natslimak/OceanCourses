**Valid (you CAN say without calibration):**  
"There are more fish here than there" — you can compare one part of the echogram to another just by looking at which area is darker/brighter.

**Invalid (you CANNOT say without calibration):**  
"There are 500 tons of fish" or "the backscatter is −40 dB" — you need calibration to know the actual numbers because without it, you don't know if your system is sensitive or weak.

**Why?** Calibration tells you: *"1 unit of echo = X amount of fish."* Without it, the numbers are meaningless.

**Why the top is hard to interpret:**  
The transducer's near-field creates a "blanking zone" where the beam is still forming and is very narrow/intense. Echoes from this zone saturate the receiver or are too strong to process accurately, and sound hasn't fully spread yet, so you get artifacts and noise instead of clean fish signals.

**Instruments setting/Physical factors that controls this zone size:**  
- Pulse length (transmission duration) — longer pulses create a larger blanking zone
- Transducer frequency — lower frequencies have larger near-fields
- Transducer diameter/size — larger transducers have larger near-fields
- Receiver gain — higher gain amplifies saturation effects in the near-field
- Transmit power — higher power increases saturation in the blanking zone

**The Problem:**
In single-beam sonar, if a fish is off to the side (not directly below), it receives less sound energy and reflects a weaker echo. This makes the fish appear smaller than it really is.

**The Solution (Split-beam):**
Split-beam systems divide the beam into sections and measure which section the echo came from. They then apply a correction factor to account for the fish being off-center, restoring the true echo strength.

**Why It Matters:**
- **Fish sizing:** Accurate echo strength → accurate length estimate (via TS-length equations)
- **Density estimation:** Fewer false negatives (small fish missed) and more accurate counts → better abundance estimates

**In 2-3 sentences for your answer:**
Split-beam angle estimation corrects for beam-pattern losses by identifying where in the beam each target is located and adjusting the target strength accordingly. This eliminates the artificial weakening of off-axis targets, ensuring accurate fish length estimates from TS values. Better TS accuracy leads to reliable fish density and abundance estimates.

**Why Bubbles Create a Strong Surface Layer:**
Bubbles have a gas-water interface with very different acoustic impedance, causing strong reflection. A bubble layer near the surface acts like an acoustic mirror, reflecting sound back toward the source.

**Why Detection is Reduced at Depth:**
The strong surface reflection bounces energy back up, so less acoustic energy penetrates downward to reach deeper targets. This creates a "shadow zone" below the bubble layer.

**Bubble Scattering vs. Bubble Attenuation:**
- **Scattering:** Bubbles redirect/deflect sound in all directions (reflection, diffusion)
- **Attenuation:** Bubbles absorb acoustic energy as sound passes through the bubble layer, weakening the signal

**Frequency Dependence:**
Bubble effects worsen at **lower frequencies** because the bubble resonance frequency is lower, causing stronger absorption and scattering. Higher frequencies penetrate bubble layers better but are still affected.

**In 2-3 sentences for your answer:**
Bubbles create a strong acoustic interface that reflects energy back and attenuates downward-traveling sound, reducing penetration to deeper fish. Bubble scattering is directional deflection, while attenuation is energy absorption; both effects are worst at low frequencies where bubble resonance is strongest.

- **Sound speed error → mostly biases depth.**  
  If the assumed sound speed is wrong, the time‑to‑range conversion is wrong, shifting all depth estimates.

- **Absorption (attenuation) error → mostly biases TS and \(S_v\).**  
  If absorption is underestimated, echoes look too strong with range, inflating TS and volume backscatter.

- **Beamwidth/beam angle error → mostly biases TS and \(S_v\).**  
  Wrong beam pattern correction makes off‑axis targets appear too weak/strong.

- **Bottom picking error → mostly biases depth (and can affect \(S_v\) near bottom).**  
  A mis‑picked bottom shifts the seafloor line, misplacing target depths and contaminating near‑bottom backscatter.

When sound speed increases with depth, rays bend **upward toward the slower, shallower water**. If you assume straight‑line paths, you’ll **misplace targets**—you’ll estimate **too deep and too far out** (angle and range bias) because the actual curved path is shorter/shallower than the straight‑line assumption.

# Part 2: Split Beam

```python
Depth = 50      # distance to target in meters
Offset = 5      # horizontal offset in meters

angle = math.degrees(math.atan(Offset / Depth))  # calculate angle in degrees
print(f"Angle to target: {angle:.2f} degrees")
```

```python
d = 0.331 # distance between two transducers in meters

R = np.sqrt(Depth**2 + Offset**2)  
R_1 = np.sqrt(Depth**2 + (Offset+d/2)**2)
R_2 = np.sqrt(Depth**2 + (Offset-d/2)**2)
delta_R = R_1 - R_2
print(f"Path difference (delta R): {delta_R:.4f} m")
```

```python
c = 1500            # speed of sound in water in m/s
f = 18 * 1000       # frequency in Hz (18 kHz)

wavelength = c / f  # wavelength in meters
print(f"Wavelength: {wavelength:.4f} m")
```

```python
phase_shift = (2 * math.pi * f /  c) * delta_R
print(f"Phase shift (delta phi): {math.degrees(phase_shift):.4f} degrees")
```

**What phase shift means:**

The **phase shift** tells you how "out of sync" the two received echo signals are.

**Physical explanation:**
- The target is **off to one side** (5 m offset)
- Sound travels **different distances** to reach the left vs. right element centers
- The path difference is **ΔR ≈ 0.033 m**
- Since one wave travels 0.033 m farther, it arrives **later in its oscillation cycle**

**What 143° means:**
- One complete wave cycle = 360°
- 143° means the signals are **about 40% of a cycle out of phase** (143/360 ≈ 0.4)
- If you plotted both signals, one would be **shifted horizontally** by about 40% of a wavelength
- The split-beam system **measures this phase difference** to calculate the target's angle

**Why it matters:**
The echosounder compares phases between element pairs (like 1+2 vs 3+4) to determine **where the target is** relative to the beam axis—that's how it gets precise angle measurements beyond just "which sector detected it."

```python
frequencies = [38*1000, 70*1000, 120*1000, 200*1000, 330*1000]

for freq in frequencies:
    ratio = f/freq
    d_new = d * ratio
    print(f"Frequency: {freq} kHz, New distance between transducers: {d_new:.3f} m")
```

```python
f18 = 18e3              # 18 kHz
f38 = 38e3              # 38 kHz

theta18_deg = 5.5       # half beamwidth at 18 kHz
theta38_deg = 3.5       # half beamwidth at 38 kHz

d18 = 0.331             # center spacing at 18 kHz [m]

# Convert degrees to radians
theta18 = np.deg2rad(theta18_deg)
theta38 = np.deg2rad(theta38_deg)

# -------------------------
# Solve for d38
# Using:
# (2π f18 / c) * d18 * sin(theta18)
# =
# (2π f38 / c) * d38 * sin(theta38)
# -------------------------

d38 = d18 * (f18 / f38) * (np.sin(theta18) / np.sin(theta38))

print(f"Required center distance at 38 kHz: {d38:.3f} m")
```
