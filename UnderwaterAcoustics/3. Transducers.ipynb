


Range resolution = The minimum distance (in depth/range) between two objects so that you can tell them apart as separate objects rather than one blob.

Range resolution is specifically about separating things vertically/in depth, not about coverage area. Think of it as the sonar's ability to say: "These are TWO different things, not one thing." 



ANSWER:

(i) Range resolution decreases (gets worse) - You can't distinguish between two objects that are closer together in range.
(ii) Visible change in echogram for thin layers - Thin sediment layers will appear thicker/broader in the echogram because the longer pulse "smears out" the return signal, making it harder to resolve fine vertical structure.

```python
initial_tao = 0.0001            # initial pulse duration in seconds  (SHORT PULSE)
increased_tao = 0.001           # increased pulse duration in seconds (LONG PULSE)
c = 1500                        # speed of sound in water in m/s

R_initial = 0.5 * c * initial_tao       # range in meters (assuming sound speed = 1500 m/s)
R_increased = 0.5 * c * increased_tao   # range in meters (assuming sound speed = 1500 m/s)

print(f"Initial range: {R_initial*100:.2f} cm")
print(f"Increased range: {R_increased*100:.2f} cm")
```
